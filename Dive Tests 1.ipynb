```python
import csv;
import numpy as np;
import matplotlib;
import matplotlib.pyplot as plt;


# physical properties
g = 9.81;            # m/s^2
rho = 1.225;         # kg/m^3
m = 1.38;            # lbm
m = 4.44822162*m/g;  # convert to kg
A = 0.103;           # m^2
flight_time = 253; # seconds

# load data
file ="raw_data.csv";
data = np.loadtxt(open(file, "rb"), delimiter=",", skiprows=1, usecols = (1,3));
data_point_count = len(data[:,0]);


# data sets
a = data[:,1]*g;      # m/s^2
V = data[:,0]/3.6;  # m/s (from kph)
t = np.linspace(0,flight_time,num=data_point_count);  # s
CD = -m*(a-g)/(A*0.5*rho*np.power(V,2));   # drag coefficient in dive

range_1 = [63, 65.5]
range_2 = [92.5, 95]
range_3 = [121, 123.5]
range_4 = [171.5, 174]
range_5 = [200, 204]

view_range_1 = [58, 70]
view_range_2 = [90, 97]
view_range_3 = [117, 127]
view_range_4 = [169, 175]
view_range_5 = [196, 210]


i_1 = [idx for idx,val in enumerate(t) if range_1[0] < val and val < range_1[1]]
i_2 = [idx for idx,val in enumerate(t) if range_2[0] < val and val < range_2[1]]
i_3 = [idx for idx,val in enumerate(t) if range_3[0] < val and val < range_3[1]]
i_4 = [idx for idx,val in enumerate(t) if range_4[0] < val and val < range_4[1]]
i_5 = [idx for idx,val in enumerate(t) if range_5[0] < val and val < range_5[1]]


################
#### DIVE 1 ####
################

fig1, pL = plt.subplots();
plt.plot(t,a,'b')
plt.plot(t,V,'g')
pL.set_ylim(-10,40)
plt.ylabel('airspeed (m/s) and x-accel (m/s^2)')
plt.xlabel('flight time (s)')

pR = plt.twinx()
plt.plot(t,CD,'r')
plt.ylabel('CD')
pR.set_ylim(0,0.5)
pR.set_xlim(view_range_1[0],view_range_1[1])
plt.axvspan(range_1[0], range_1[1], color='green', alpha=0.3)
fig1.legend(['airspeed','x-accel','CD','Drag calc'],loc = [0.15,0.65])
plt.title(label= 'Dive 1')

#fig1 = plt.gcf()
#fig1.set_size_inches(fig_width, fig_height)
#w = plt.get_current_fig_manager() 
#w.window.wm_geometry("+"+str(fig1_loc[0]*screen_width)+"+"+str(fig1_loc[1]*screen_height))

################
#### DIVE 2 ####
################

fig2, pL = plt.subplots();
plt.plot(t,a,'b')
plt.plot(t,V,'g')
pL.set_ylim(-10,40)
plt.ylabel('airspeed (m/s) and x-accel (m/s^2)')
plt.xlabel('flight time (s)')

pR = plt.twinx()
plt.plot(t,CD,'r')
plt.ylabel('CD')
pR.set_ylim(0,0.5)
pR.set_xlim(view_range_2[0],view_range_2[1])
plt.axvspan(range_2[0], range_2[1], color='green', alpha=0.3)
fig2.legend(['airspeed','x-accel','CD','Drag calc'],loc = [0.15,0.65])
plt.title(label= 'Dive 2')

#fig1 = plt.gcf()
#fig1.set_size_inches(fig_width, fig_height)
#w = plt.get_current_fig_manager() 
#w.window.wm_geometry("+"+str(int(fig2_loc[0]*screen_width))+"+"+str(int(fig2_loc[1]*screen_height)))


################
#### DIVE 3 ####
################

fig3, pL = plt.subplots();
plt.plot(t,a,'b')
plt.plot(t,V,'g')
pL.set_ylim(-10,40)
plt.ylabel('airspeed (m/s) and x-accel (m/s^2)')
plt.xlabel('flight time (s)')

pR = plt.twinx()
plt.plot(t,CD,'r')
plt.ylabel('CD')
pR.set_ylim(0,0.5)
pR.set_xlim(view_range_3[0],view_range_3[1])
plt.axvspan(range_3[0], range_3[1], color='green', alpha=0.3)
fig3.legend(['airspeed','x-accel','CD','Drag calc'],loc = [0.15,0.65])
plt.title(label= 'Dive 3')

#fig3 = plt.gcf()
#fig3.set_size_inches(fig_width, fig_height)
#w = plt.get_current_fig_manager() 
#w.window.wm_geometry("+"+str(int(fig3_loc[0]*screen_width))+"+"+str(int(fig3_loc[1]*screen_height)))


################
#### DIVE 4 ####
################

fig4, pL = plt.subplots();
plt.plot(t,a,'b')
plt.plot(t,V,'g')
pL.set_ylim(-10,40)
plt.ylabel('airspeed (m/s) and x-accel (m/s^2)')
plt.xlabel('flight time (s)')

pR = plt.twinx()
plt.plot(t,CD,'r')
plt.ylabel('CD')
pR.set_ylim(0,0.5)
pR.set_xlim(view_range_4[0],view_range_4[1])
plt.axvspan(range_4[0], range_4[1], color='green', alpha=0.3)
fig4.legend(['airspeed','x-accel','CD','Drag calc'],loc = [0.15,0.65])
plt.title(label= 'Dive 4')

#fig4 = plt.gcf()
#fig4.set_size_inches(fig_width, fig_height)
#w = plt.get_current_fig_manager()
#w.window.wm_geometry("+"+str(int(fig4_loc[0]*screen_width))+"+"+str(int(fig4_loc[1]*screen_height)))

################
#### DIVE 5 ####
################

fig5, pL = plt.subplots();
plt.plot(t,a,'b')
plt.plot(t,V,'g')
pL.set_ylim(-10,40)
plt.ylabel('airspeed (m/s) and x-accel (m/s^2)')
plt.xlabel('flight time (s)')

pR = plt.twinx()
plt.plot(t,CD,'r')
plt.ylabel('CD')
pR.set_ylim(0,0.5)
pR.set_xlim(view_range_5[0],view_range_5[1])
plt.axvspan(range_5[0], range_5[1], color='green', alpha=0.3)
fig5.legend(['airspeed','x-accel','CD','Drag calc'],loc = [0.15,0.65])
plt.title(label= 'Dive 5')

#fig5 = plt.gcf()
#fig5.set_size_inches(fig_width, fig_height)
#w = plt.get_current_fig_manager()
#w.window.wm_geometry("+"+str(int(fig5_loc[0]*screen_width))+"+"+str(int(fig5_loc[1]*screen_height)))


################
#### Drag Curve ####
################

fig6, pL = plt.subplots();
plt.plot(V[i_1],CD[i_1],'r.')
plt.plot(V[i_2],CD[i_2],'g.')
plt.plot(V[i_3],CD[i_3],'b.')
plt.plot(V[i_4],CD[i_4],'k.')
plt.plot(V[i_5],CD[i_5],'c.')
pL.set_ylim(0,0.8)
pL.set_xlim(10,30)
plt.ylabel('CD')
plt.xlabel('airspeed (m/s)')

fig6.legend(['dive 1','dive 2','dive 3','dive 4','dive 5',],loc = [0.7,0.6])
plt.title(label= 'CD vs airspeed')

#fig6 = plt.gcf()
#fig6.set_size_inches(fig_width, fig_height)
#w = plt.get_current_fig_manager()
#w.window.wm_geometry("+"+str(int(fig6_loc[0]*screen_width))+"+"+str(int(fig6_loc[1]*screen_height)))





plt.show()

# csvfile = open(file,newline='');
# csv_reader = csv.reader(csvfile, delimiter=',')
# line_count = 0

# x = list(csv_reader)
# result = numpy.array(x).astype("float")

#    for row in csv_reader:
#        print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
#        line_count += 1
#    print(f'Processed {line_count} lines.')
```
